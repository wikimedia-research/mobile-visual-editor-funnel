# Attempts and completion by interface and experience

```python
attempts = hive.run("""
-- Transform the steps into attempts
with steps as (
    select
        event.editing_session_id as attempt_id,
        unix_timestamp(
            dt,
            "yyyy-MM-dd'T'hh:mm:ssX"
        ) as ts,
        wiki,
        database_group,
        event.platform as platform,
        event.editor_interface as interface,
        event.action as action,
        event.is_oversample as is_oversample,
        event.user_editcount as user_edit_count,
        coalesce(event.user_class = "IP", False) as user_is_unregistered
    from event_sanitized.editattemptstep
    left join canonical_data.wikis
    on wiki = database_code
    where
        -- Remove Flow and other non-standard edits
        event.integration = "page" and
        -- Remove events not needed to calculate session starts or completions
        event.action in ("ready", "saveSuccess") and
        -- One month of data
        year = 2019 and (
            (month = 2 and day >= 8) or
            (month = 9)
        )
    -- Ensure that steps within each session are in chronological order without having to do a global order by
    distribute by attempt_id
    sort by ts asc
)
-- Group those steps into attempts
select
    attempt_id,
    from_unixtime(min(ts), "yyyy-MM-dd'T'hh:mm:ssX") as dt,
    min(wiki) as wiki,
    case
        when min(database_group) != "wikipedia" then Null
        when min(wiki) in ("enwiki", "nlwiki", "eswiki") then False
        else True
    end as ve_default_wiki,
    min(platform) as platform,
    -- Select the last editor in the session as the "real" one
    collect_list(interface)[
        size(collect_list(interface)) - 1
    ] as editor,
    sum(cast(action = "saveSuccess" as int)) >= 1 as is_completed,
    max(is_oversample) as is_oversample,
    case
        when max(user_is_unregistered) or max(user_edit_count) is null then "unregistered"
        when max(user_edit_count) = 0 then "0"
        when max(user_edit_count) < 10 then "1-9"
        when max(user_edit_count) < 100 then "10-99"
        when max(user_edit_count) < 1000 then "100-999"
        else "1000+"
    end as user_experience
from steps        
group by attempt_id
""").assign(
    dt=lambda df: pd.to_datetime(df["dt"]),
    wiki=lambda df: pd.Categorical(df["wiki"]),
    platform=lambda df: pd.Categorical(df["platform"]),
    editor=lambda df: pd.Categorical(df["editor"]),
    user_experience=lambda df: pd.Categorical(
        df["user_experience"],
        categories=["unregistered", "0", "1-9", "10-99", "100-999", "1000+"], ordered=True
    )
)
```

## Average daily edit attempts

We consider any session that reaches the `ready` event as an attempt, to filter out what seems to be a large number of bot inits (https://phabricator.wikimedia.org/T209995).

```python
daily_attempts = (
    attempts
    .query("~is_oversample")
    .pivot_table(
        columns=["platform", "editor"],
        index="user_experience",
        values="dt",
        aggfunc="count",
        margins="True",
        margins_name="total"
    )
    .multiply((16/30)) # Correct for sampling, reduce to daily average
)

(
    daily_attempts
    .style
    .background_gradient(
        axis=None, 
        subset=(~daily_attempts.index.str.match("total"), ~daily_attempts.columns.get_level_values(0).str.match("total"))
    )
    .format(lambda x: num_str(int(x)))
)
```

## Completion rates

```python
completion = (
    attempts
    .query("~is_oversample & ~(editor == 'wikitext-2017')")
    .assign(user_experience=lambda df: df["user_experience"].cat.as_unordered()) # Work around https://github.com/pandas-dev/pandas/issues/25815
    .pivot_table(
        columns=["platform", "editor"],
        index="user_experience",
        values="is_completed",
        aggfunc=lambda grp: grp.sum() / len(grp),
        margins="True",
        margins_name="total"
    )
)

(
    completion
    .style
    .background_gradient(
        axis=None, 
        subset=(~completion.index.str.match("total"), ~completion.columns.get_level_values(0).str.match("total"))
    )
    .format("{:.0%}")
)
```

# Desktop completion rate by interface, experience, and VE default status

```python
desktop_completion = (
    attempts
    .query("platform == 'desktop' & editor == 'visualeditor' & ~ve_default_wiki.isnull()")
    .pivot_table(
        columns=["ve_default_wiki", "editor"],
        index="user_experience",
        values="is_completed",
        aggfunc=lambda grp: grp.sum() / len(grp),
        margins="True",
        margins_name="total"
    )
)

desktop_completion.insert(2, "wikitext", completion[("desktop", "wikitext")])

desktop_completion = (
    desktop_completion
    .set_axis([
        "visualeditor (non-default wiki)", 
        "visualeditor (default wiki)", 
        "wikitext", 
        "total"
    ], axis=1, inplace=False)
)

(
    desktop_completion
    .style
    .background_gradient(
        axis=None,
        subset=(~desktop_completion.index.str.match("total"), ~desktop_completion.columns.get_level_values(0).str.match("total"))
    )
    .format("{:.0%}")
)
```

# Mobile visual editor edit completion rate

```python
completion_over_time_r = hive.run("""
with attempts as (
    select
        to_date(min(dt)) as date,
        sum(cast(event.action = "saveSuccess" as int)) >= 1 as completed,
        min(event.user_editcount) between 0 and 100 as user_is_inexperienced
    from event_sanitized.editattemptstep
    where
        event.platform = "phone" and
        event.editor_interface = "visualeditor" and
        event.action in ("ready", "saveSuccess") and
        year > 0
    group by event.editing_session_id
)
select
    date,
    sum(cast(completed as int)) / count(*) as edit_completion,
    sum(cast(completed and user_is_inexperienced as int)) / sum(cast(user_is_inexperienced as int)) as inexperienced_user_edit_completion
from attempts
group by date
""")
```

```python
completion_over_time = (
    completion_over_time_r
    .assign(date=lambda df: pd.to_datetime(df["date"]))
    .set_index("date")
    # Drop bad data due to T221197#5305019
    .drop(pd.date_range(start="2019-06-26", end="2019-07-03"))
    # Insert NaNs for missing dates
    .asfreq("D")
    .rename({
        "edit_completion": "overall",
        "inexperienced_user_edit_completion": "inexperienced editors (0-99 edits)"
    }, axis=1)
    .rolling(7).mean()
    .query("date < '2019-07'")
)
```

```python
ax = completion_over_time.plot(ylim=(0, 0.35))
plt.ylabel("edit attempts leading to published edit")
plt.xlabel(None)
ax.yaxis.set_major_formatter(charting.pct_fmt)
```

# Mobile VE edits over time

```python
mobile_edit_trends_r = hive.run("""
select
    to_date(event_timestamp) as date,
    sum(cast(not array_contains(revision_tags, "visualeditor") as int)) as wt_edits,
    sum(cast(array_contains(revision_tags, "visualeditor") as int)) as ve_edits,
    (
        sum(cast(revision_is_identity_reverted and array_contains(revision_tags, "visualeditor") as int)) / 
        sum(cast(array_contains(revision_tags, "visualeditor") as int))
    ) as ve_revert_rate,
    (
        sum(cast(revision_is_identity_reverted and not array_contains(revision_tags, "visualeditor") as int)) / 
        sum(cast(not array_contains(revision_tags, "visualeditor") as int))
    ) as wt_revert_rate
from wmf.mediawiki_history
where
    array_contains(revision_tags, "mobile web edit") and
    event_timestamp > "2018" and
    event_entity = "revision" and 
    snapshot = "2019-06"
group by to_date(event_timestamp)
""")
```

```python
mobile_edit_trends = (
    mobile_edit_trends_r
    .assign(date=lambda df: pd.to_datetime(df["date"]))
    .sort_values("date")
    .set_index("date")
    .query("date < '2019-07'")
    .rolling(7).mean()
    .dropna(thresh=4)
)

mobile_edit_trends.head()
```

```python
ax = mobile_edit_trends["ve_edits"].plot(ylim=(0, 5500))
plt.title("Edits published using the mobile visual editor")
plt.ylabel("edits per day")
plt.xlabel(None);
```

```python
# Percent change over the fiscal year
base_edits = mobile_edit_trends.loc["2018-06", "ve_edits"].mean()
cur_edits = mobile_edit_trends.loc["2019-06", "ve_edits"].mean()

print(pct_str(cur_edits / base_edits - 1))
```

```python
daily_edits = mobile_edit_trends[["ve_edits", "wt_edits"]]

edits_index = (
    (daily_edits / daily_edits.iloc[0])
    .rename({
        "ve_edits": "visual editor",
        "wt_edits": "wikitext editor"
    }, axis=1)
)

ax = edits_index.plot(ylim=(0.8, 2.6))
plt.title("mobile web edits by editing interface")
plt.ylabel("daily edits as % of Jan 2018 value")
plt.xlabel(None)
ax.yaxis.set_major_formatter(charting.pct_fmt)
```

# Revert rates

```python
revert_rates_r = hive.run("""
select
    to_date(event_timestamp) as date,
    (
        sum(cast(
            revision_is_identity_reverted and 
            revision_seconds_to_identity_revert <= 172800 and -- 48 hours
            array_contains(revision_tags, "visualeditor")
        as int)) / 
        sum(cast(array_contains(revision_tags, "visualeditor") as int))
    ) as `visual editor`,
    (
        sum(cast(
            revision_is_identity_reverted and 
            revision_seconds_to_identity_revert <= 172800 and -- 48 hours
            not array_contains(revision_tags, "visualeditor")
        as int)) / 
        sum(cast(not array_contains(revision_tags, "visualeditor") as int))
    ) as `wikitext editor`
from wmf.mediawiki_history
where
    array_contains(revision_tags, "mobile web edit") and
    -- Avoid censoring of data
    event_timestamp between "2018" and "2019-06-28" and
    event_entity = "revision" and 
    snapshot = "2019-06"
group by to_date(event_timestamp)
""")
```

```python
revert_rates = (
    revert_rates_r
    .assign(date=lambda df: pd.to_datetime(df["date"]))
    .sort_values("date")
    .set_index("date")
    .rolling(7).mean()
    .dropna(thresh=2)
)

revert_rates.head()
```

```python
ax = revert_rates["visual editor"].plot(ylim=(0, 0.45))
plt.title("revert rate of mobile visual edits")
plt.ylabel("% of edits reverted")
plt.xlabel(None)
ax.yaxis.set_major_formatter(charting.pct_fmt);
```

```python
# Percent change over the fiscal year
base_revert_rate = revert_rates.loc["2018-06", "visual editor"].mean()
cur_revert_rate = revert_rates.loc["2019-06", "visual editor"].mean()

print(pct_str(cur_revert_rate / base_revert_rate - 1))
```

```python
cur_revert_rate
```

```python
revert_index = (revert_rates / revert_rates.iloc[0])

ax = revert_index.plot(ylim=(0))
plt.title("change in revert rate by editing interface")
plt.ylabel("revert rate as % of Jan 2018 value")
plt.xlabel(None)
ax.yaxis.set_major_formatter(charting.pct_fmt);
```

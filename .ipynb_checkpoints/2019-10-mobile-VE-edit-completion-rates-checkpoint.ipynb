```r
shhh <- function(expr) suppressPackageStartupMessages(suppressWarnings(suppressMessages(expr)))
shhh({
   library(tidyverse); library(glue); library(lubridate); library(scales); library(zoo)
})
```

# Mobile VE Edit Completion Rate -- Registered Users

```r
query <- 
"with attempts as (
    select
        to_date(min(dt)) as date,
        sum(cast(event.action = 'saveSuccess' as int)) >= 1 as completed,
        min(event.user_editcount) between 1 and 4 as one_to_four_edits,
        min(event.user_editcount)  >= 5 AND min(event.user_editcount) < 100 as five_99_edits,
        min(event.user_editcount)  >= 100 AND min(event.user_editcount) < 1000 as onehundred_999_edits,
        min(event.user_editcount)  >= 1000 as over_1000
    from event_sanitized.editattemptstep
    inner join canonical_data.wikis 
    on wiki = database_code and
    database_group ==  'wikipedia'
    where
        event.platform = 'phone' and
        event.editor_interface = 'visualeditor' and
        event.action in ('ready', 'saveSuccess') and
        event.user_id !=0 and --remove unregistered users
        year = 2019 and ((month = 02 and day >= 08) or (month >= 03 and month <= 09))  
    group by event.editing_session_id
)
select
    date,
    sum(cast(completed as int)) / count(*) as edit_completion,
    sum(cast(completed and one_to_four_edits as int)) / sum(cast(one_to_four_edits as int)) as one_to_four_edits,
    sum(cast(completed and five_99_edits as int)) / sum(cast(five_99_edits as int)) as five_99_edits,
    sum(cast(completed and onehundred_999_edits as int)) / sum(cast(onehundred_999_edits as int)) as onehundred_999_edits,
    sum(cast(completed and over_1000 as int)) / sum(cast(over_1000 as int)) as over_1000_edits
from attempts
group by date"
```

```r
completion_over_time <- wmf::query_hive(query)
```

```r
completion_over_time$date <- as.Date(completion_over_time$date, format = "%Y-%m-%d")
```

```r
#Remove bad data due to https://phabricator.wikimedia.org/T221197#5305019

completion_over_time <- subset(completion_over_time, date < "2019-06-26" | date > "2019-07-03")
```

## Overall Trend

```r
tail(completion_over_time)
```

```r
# rename columns
names(completion_over_time) <- c("date", "overall", "1-4 edits", "5-99 edits", "100-999 edits", "1,000+ edits")
```

```r
p <- completion_over_time %>%
  gather("editor_experience", "edit_attempts", 2:6)  %>%
  mutate(rolling_average = rollmean(as.numeric(edit_attempts), 7, na.pad=TRUE, align="right")) %>%
  ggplot(aes(x= date, y = rolling_average, color = editor_experience, alpha = editor_experience)) +    
  geom_line(size = 1.5) +
  scale_y_continuous("edit attempts leading to published edit", limits=c(0,.80), labels = scales::percent) +
  scale_x_date("Date", labels = date_format("%b %Y"), date_breaks = "1 month") +
  labs(title = "Mobile visual editor edit completion rates \n by editor experience for all registered users") +
  ggthemes::theme_tufte(base_size = 14, base_family = "Gill Sans") +
    scale_color_brewer(palette = 'Set1') +
    scale_alpha_manual(values=c("0.4", "0.4", "0.4", "0.4", "1")) +
    theme(axis.text.x=element_text(angle = 45, hjust = 1),
        plot.title = element_text(hjust = 0.5),
        panel.grid = element_line("gray70"),
        legend.position = "right",
        legend.title  = element_text(size=12, face = 'bold'),
        legend.text = element_text(size=12)
    )
p
ggsave("Figures/mobile_ve_editcompletion_registered.png", p, width = 18, height = 9, units = "in", dpi = 150)
```

### Average completion rate for first month of full data

# Mobile VE Edit Completion Rate -- All Users

```r
query <- 
"with attempts as (
    select
        to_date(min(dt)) as date,
        sum(cast(event.action = 'saveSuccess' as int)) >= 1 as completed,
        min(event.user_id) = 0 as unregistered,
        min(event.user_editcount) between 1 and 4 as one_to_four_edits,
        min(event.user_editcount)  >= 5 AND min(event.user_editcount) < 100 as five_99_edits,
        min(event.user_editcount)  >= 100 AND min(event.user_editcount) < 1000 as onehunderd_999_edits,
        min(event.user_editcount)  >= 1000 as over_1000
    from event_sanitized.editattemptstep
    inner join canonical_data.wikis 
    on wiki = database_code and
    database_group ==  'wikipedia'
    where
        event.platform = 'phone' and
        event.editor_interface = 'visualeditor' and
        event.action in ('ready', 'saveSuccess') and
        year = 2019 and ((month = 02 and day >= 08) or (month >= 03 and month <= 09))  
    group by event.editing_session_id
)
select
    date,
    sum(cast(completed as int)) / count(*) as edit_completion,
    sum(cast(completed and unregistered as int)) / sum(cast(unregistered as int)) as unregistered,
    sum(cast(completed and one_to_four_edits as int)) / sum(cast(one_to_four_edits as int)) as one_to_four_edits,
    sum(cast(completed and five_99_edits as int)) / sum(cast(five_99_edits as int)) as five_99_edits,
    sum(cast(completed and onehunderd_999_edits as int)) / sum(cast(onehunderd_999_edits as int)) as onehunderd_999_edits,
    sum(cast(completed and over_1000 as int)) / sum(cast(over_1000 as int)) as over_1000_edits
from attempts
group by date"
```

```r
completion_over_time_incl_nonregistered <- wmf::query_hive(query)
```

```r
completion_over_time_incl_nonregistered$date <- as.Date(completion_over_time_incl_nonregistered$date, format = "%Y-%m-%d")
```

```r
#Remove bad data due to https://phabricator.wikimedia.org/T221197#5305019

completion_over_time_incl_nonregistered <- subset(completion_over_time_incl_nonregistered, date < "2019-06-26" | date > "2019-07-03")
```

```r
# rename columns
names(completion_over_time_incl_nonregistered) <- c("date", "overall", "unregistered", "1-4 edits", "5-99 edits", "100-999 edits", "1,000+ edits")
```

```r
tail(completion_over_time_incl_nonregistered)
```

## Overall Trend

```r
p <- completion_over_time_incl_nonregistered %>%
  gather("editor_experience", "edit_attempts", 2:7)  %>%
   mutate(rolling_average = rollmean(as.numeric(edit_attempts), 7, na.pad=TRUE, align="right")) %>%
  ggplot(aes(x= date, y = rolling_average, color = editor_experience, alpha = editor_experience)) +    
  geom_line(size = 1.5) +
  scale_y_continuous("edit attempts leading to published edit", limits=c(0,.80), labels = scales::percent) +
  scale_x_date("Date", labels = date_format("%b %Y"), date_breaks = "1 month") +
  labs(title = "Mobile visual editor edit completion rates \n by editor experience for all users") +
  ggthemes::theme_tufte(base_size = 14, base_family = "Gill Sans") +
    scale_color_brewer(palette = 'Dark2') +
    scale_alpha_manual(values=c("0.4", "0.4", "0.4", "0.4", "1", "0.4")) +
    theme(axis.text.x=element_text(angle = 45, hjust = 1),
        plot.title = element_text(hjust = 0.5),
        panel.grid = element_line("gray70"),
         legend.position = 'right',
        legend.title  = element_text(size=12, face = 'bold'),
        legend.text = element_text(size=12))
p

ggsave("Figures/mobile_ve_editcompletion_all.png", p, width = 18, height = 9, units = "in", dpi = 150)
```

There is a 10% drop in the overall rate and unregistred editors that occurs on July 4th and brings the overall number of editors down as well. Further investigation needed to determine cause.

## Pre and Post Edit Card deployment Analysis


```r
#Pre edit cards completion rate

pre_edit_card_completion_rate_all <- completion_over_time_incl_nonregistered  %>%
    filter(date >= "2019-05-01" & date <= '2019-05-31') %>%
    gather("editor_experience", "edit_attempts", 2:7) 

p <- pre_edit_card_completion_rate_all %>%
   #mutate(rolling_average = rollmean(as.numeric(edit_attempts), 7, na.pad=TRUE, align="right")) %>%
  ggplot(aes(x= date, y = as.numeric(edit_attempts), color = editor_experience, alpha = editor_experience)) +    
  geom_line(size = 1) +
  scale_y_continuous("edit attempts leading to published edit", limits=c(0,.80), labels = scales::percent) +
  scale_x_date("Date", labels = date_format("%b %d %Y"), date_breaks = "1 week") +
  labs(title = "Pre-edit card deployment mobile VE completion rates \n by editor experience",
      caption = "Includes both logged in and logged out users") +
  ggthemes::theme_tufte(base_size = 14, base_family = "Gill Sans") +
  scale_color_brewer(palette = 'Dark2') +
    scale_alpha_manual(values=c("0.4", "0.4", "0.4", "0.4", "1", "0.4")) +
    theme(axis.text.x=element_text(angle = 45, hjust = 1),
        plot.title = element_text(hjust = 0.5),
        panel.grid = element_line("gray70"),
         legend.position = 'right',
        legend.title  = element_text(size=12, face = 'bold'),
        legend.text = element_text(size=12))

p

ggsave("Figures/pre_edit_card_editcompletion_all.png", p, width = 18, height = 9, units = "in", dpi = 150)

```

### Average Pre Edit Card Deployment Edit Completion Rate (May 2019)

```r
pre_edit_card_completion_rate_avg_all <- pre_edit_card_completion_rate_all %>%
    group_by(editor_experience) %>%
    summarise(pre_avg_completion_rate = mean(as.numeric(edit_attempts)))

pre_edit_card_completion_rate_avg
```

```r
#Post edit cards completion rate

post_edit_card_completion_rate_all <- completion_over_time_incl_nonregistered %>%
    filter(date >= "2019-09-01" & date <= '2019-09-30') %>%
    gather("editor_experience", "edit_attempts", 2:7) 

p <- post_edit_card_completion_rate_all %>%
   #mutate(rolling_average = rollmean(as.numeric(edit_attempts), 7, na.pad=TRUE, align="right")) %>%
  ggplot(aes(x= date, y = as.numeric(edit_attempts), color = editor_experience, alpha = editor_experience)) +    
  geom_line(size = 1) +
  scale_y_continuous("edit attempts leading to published edit", limits=c(0,.80), labels = scales::percent) +
  scale_x_date("Date", labels = date_format("%b %d %Y"), date_breaks = "1 week") +
  labs(title = "Post-edit card deployment mobile VE completion rates \n by editor experience",
    caption = "Includes both logged in and logged out users") +
  ggthemes::theme_tufte(base_size = 14, base_family = "Gill Sans") +
    scale_color_brewer(palette = 'Dark2') +
    scale_alpha_manual(values=c("0.4", "0.4", "0.4", "0.4", "1", "0.4")) +
    theme(
        plot.title = element_text(hjust = 0.5),
        panel.grid = element_line("gray70"),
         legend.position = 'right') 

p
ggsave("Figures/post_edit_card_editcompletion_all.png", p, width = 18, height = 9, units = "in", dpi = 150)
```
